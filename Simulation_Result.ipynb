```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
regret_cumule_cklucb = np.load('./Sim_Data/cklucb_regret.npy')
regret_cumule_klucb = np.load('./Sim_Data/klucb_regret.npy')
regret_cumule_c_ts = np.load('./Sim_Data/cts_regret.npy')
regret_cumule_ts = np.load('./Sim_Data/ts_regret.npy')
```

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(regret_cumule_c_ts[:], linestyle='-', label='C-TS')
ax1.plot(regret_cumule_ts[:], linestyle='-', label='TS')
ax1.plot(regret_cumule_cklucb[:],linestyle='--', label='C-KLUCB')
ax1.plot(regret_cumule_klucb[:],linestyle='--', label='KLUCB')

ax1.legend(loc='best')
ax1.grid()
ax1.set_title('Evolution of average cumulative regrets according to time')
plt.show()
```
