```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
regret_cumule_cklucb = np.load('./Sim_Data/cklucb_regret.npy')
regret_cumule_klucb = np.load('./Sim_Data/klucb_regret.npy')
regret_cumule_c_ts = np.load('./Sim_Data/cts_regret.npy')
regret_cumule_ts = np.load('./Sim_Data/ts_regret.npy')
```

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(regret_cumule_c_ts[:], linestyle='-', label='C-TS')
ax1.plot(regret_cumule_ts[:], linestyle='-', label='TS')
ax1.plot(regret_cumule_cklucb[:],linestyle='--', label='C-KLUCB')
ax1.plot(regret_cumule_klucb[:],linestyle='--', label='KLUCB')

ax1.legend(loc='best')
ax1.grid()
ax1.set_title('Evolution of average cumulative regrets according to time under non-stationary environment')
plt.xlabel('Time')
plt.ylabel('Regret')
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(1, 1, 1)
regret_cumule_c_ts_smooth = smooth(regret_cumule_c_ts[:],100)
regret_cumule_c_ts_smooth[-100:] = regret_cumule_c_ts_smooth[-100]
regret_cumule_ts_smooth = smooth(regret_cumule_ts[:],100)
regret_cumule_ts_smooth[-100:] = regret_cumule_ts_smooth[-100]
regret_cumule_cklucb_smooth = smooth(regret_cumule_cklucb[:],100)
regret_cumule_cklucb_smooth[-100:] = regret_cumule_cklucb_smooth[-100]
regret_cumule_klucb_smooth = smooth(regret_cumule_c_ts[:],100)
regret_cumule_c_ts_smooth[-100:] = regret_cumule_c_ts_smooth[-100]

ax1.plot(regret_cumule_c_ts_smooth, linestyle='-', label='C-TS')
ax1.plot(smooth(regret_cumule_ts[:],100), linestyle='-', label='TS')
ax1.plot(smooth(regret_cumule_cklucb[:],100),linestyle='--', label='C-KLUCB')
ax1.plot(smooth(regret_cumule_klucb[:],100),linestyle='--', label='KLUCB')

ax1.legend(loc='best')
ax1.grid()
ax1.set_title('Evolution of average cumulative regrets according to time under non-stationary environment')
plt.xlabel('Time')
plt.ylabel('Regret')
plt.show()
```
