```python
import scienceplots
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
```

```python
regret_cumule_cklucb = np.load('./Sim_Data/regret_cklucb.npy')
regret_cumule_klucb = np.load('./Sim_Data/regret_klucb.npy')
regret_cumule_c_ts = np.load('./Sim_Data/regret_c_ts.npy')
regret_cumule_ts = np.load('./Sim_Data/regret_ts.npy')
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
regret_cumule_c_ts = smooth(regret_cumule_c_ts,30)
regret_cumule_ts = smooth(regret_cumule_ts,30)
regret_cumule_cklucb = smooth(regret_cumule_cklucb,30)
regret_cumule_klucb = smooth(regret_cumule_klucb,30)
```

```python
font = font_manager.FontProperties(family='Times New Roman',
                                   weight='bold',
                                   style='normal',
                                   size=21)
```

```python
fig = plt.figure(figsize=(12, 8))
ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(regret_cumule_c_ts[:5000]-51.2, label='C-TS',linewidth=3)
ax1.plot(regret_cumule_ts[:5000]-50.1, label='TS',linewidth=3)
ax1.plot(regret_cumule_cklucb[:5000]-52.6, label='C-KLUCB',linewidth=3)
ax1.plot(regret_cumule_klucb[:5000]-52.1, label='KL-UCB',linewidth=3)
ax1.legend(loc='lower right',prop=font)
plt.tick_params(labelsize=30)
labels=ax1.get_xticklabels()+ax1.get_yticklabels()
plt.grid()
plt.xlabel('Time (s)',fontfamily='Times New Roman',fontsize=30,fontweight='bold')
plt.ylabel('Regrets',fontfamily='Times New Roman',fontsize=30,fontweight='bold')
plt.show()
```
