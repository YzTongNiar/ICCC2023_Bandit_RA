```python
import matplotlib.pyplot as plot
import numpy as np
```

```python
def kl_bernoulli(p, q):
    """
    Compute kl-divergence for Bernoulli distributions
    """
    result = p * np.log(p / q) + (1 - p) * np.log((1 - p) / (1 - q))
    return result
```

```python
p_steep = np.array([.99, .98, .96, .93, 0.90, .10, .06, .04]) #Bernoulli Parameters
p_gradual = np.array([.95, .90, .80, .65, .45, .25, .15, .10])
p_lossy = np.array([.90, .80, .70, .55, .45, .35, .20, .10])
rate = np.array([6, 9, 12, 18, 24, 36, 48, 54])
plot.figure(figsize=(12, 8))
plot.plot(p_steep,'.:')
plot.plot(p_gradual,'.:')
plot.plot(p_lossy,'.:')
plot.xlabel('Rate Index')
plot.ylabel('Success Probability')
plot.legend(['steep','gradual','lossy'])
```

```python
plot.figure(figsize=(12, 8))
plot.plot(p_steep*rate,'.:')
plot.plot(p_gradual*rate,'.:')
plot.plot(p_lossy*rate,'.:')
plot.xlabel('Rate Index')
plot.ylabel('Success Probability')
plot.legend(['steep','gradual','lossy'])
```

```python
plot.figure(figsize=(12, 8))
plot.plot(max(p_steep*rate)-p_steep*rate,'.:')
plot.plot(max(p_gradual*rate)-p_gradual*rate,'.:')
plot.plot(max(p_lossy*rate)-p_lossy*rate,'.:')
plot.xlabel('Rate Index')
plot.ylabel('Throughput Difference')
plot.legend(['steep','gradual','lossy'])
```
